Question 1: What is Simple Linear Regression (SLR)? Explain its purpose.

Answer:

*	Simple Linear Regression (SLR) is a method to find a relationship between one independent variable (X) and one dependent variable (Y).
*	It helps us predict the value of Y based on X.
*	It fits a straight line (best fit line) through the data points.
*	Purpose: To understand how one variable changes when the other changes.


Question 2: What are the key assumptions of Simple Linear Regression?

Answer:

*	There is a linear relationship between X and Y.
*	Residuals (errors) are normally distributed.
*	Constant variance (homoscedasticity) of errors.
*	Independence of errors (no correlation).
*	No extreme outliers that affect the model.


Question 3: Write the mathematical equation for a simple linear regression model and explain each term.

Answer:

*	Equation: Y = β₀ + β₁X + ε
*	Y: Dependent variable (the one we predict)
*	X: Independent variable (input)
*	β₀: Intercept (value of Y when X = 0)
*	β₁: Slope (change in Y for one unit change in X)
*	ε: Error term (difference between actual and predicted Y)


Question 4: Provide a real-world example where simple linear regression can be applied.

Answer:

*	Predicting house prices based on size (in square feet).
*	Predicting sales based on advertising budget.
*	Predicting temperature based on altitude.
*	Predicting student marks based on study hours.

Question 5: What is the method of least squares in linear regression?

Answer:

*	It’s a technique used to find the best-fitting line.
*	It minimizes the sum of the squared differences between actual and predicted Y values.
*	The line with the smallest total error is chosen as the best fit.

Question 6: What is Logistic Regression? How does it differ from Linear Regression?

Answer:

*	Logistic Regression is used when the output (Y) is categorical (like Yes/No, 0/1).
*	It predicts the probability of an event occurring.
*	It uses the sigmoid function to map output between 0 and 1.
*	Difference:
  
  o	Linear Regression → predicts continuous values.
  
  o	Logistic Regression → predicts categorical outcomes.


Question 7: Name and briefly describe three common evaluation metrics for regression models.

Answer:

*	Mean Absolute Error (MAE): Average of absolute differences between actual and predicted values.
*	Mean Squared Error (MSE): Average of squared differences between actual and predicted values.
*	R-squared (R²): Shows how well the model explains the variation in data (closer to 1 = better fit).


Question 8: What is the purpose of the R-squared metric in regression analysis?

Answer:

*	R-squared shows how much of the variation in Y is explained by X.
*	It ranges from 0 to 1.
*	Higher value = better model fit.
*	Example: R² = 0.8 means 80% of the changes in Y are explained by X.


```python
# Question 9: Write Python code to fit a simple linear regression model using scikit-learn and
# print the slope and intercept.
# Explanation :
# •	We use the LinearRegression class from scikit-learn.
# •	The model learns the best slope and intercept from data.
# •	We then print the values of both.

# Code :

# Simple Linear Regression using scikit-learn

# Step 1: Import libraries
from sklearn.linear_model import LinearRegression
import numpy as np

# Step 2: Prepare sample data
X = np.array([[1], [2], [3], [4], [5]])   # independent variable
y = np.array([2, 4, 5, 4, 5])             # dependent variable

# Step 3: Create and train the model
model = LinearRegression()
model.fit(X, y)

# Step 4: Print slope and intercept
print("Slope (β1):", model.coef_[0])
print("Intercept (β0):", model.intercept_)
```

Question 10: How do you interpret the coefficients in a simple linear regression model?

Answer:

*	Slope (β₁): Shows how much Y changes when X increases by 1 unit.

  o	Example: If β₁ = 2, Y increases by 2 when X increases by 1.

*	Intercept (β₀): Value of Y when X = 0.
*	Together, they tell how X and Y are related in the model.
